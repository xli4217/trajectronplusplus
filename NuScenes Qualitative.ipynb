```python
with open('./processed_data/nuScenes_test_mini_full.pkl', 'rb') as f:
    eval_env = dill.load(f, encoding='latin1')
eval_scenes = eval_env.scenes

# load nuscenes 
from external_libs.nuscenes.nuscenes import NuScenes
from nuscenes.eval.prediction.splits import get_prediction_challenge_split

dataroot = '/nobackup/users/xiaoli/nuscenes_dataset/dataset-mini'
nusc = NuScenes(version='v1.0-mini', dataroot=dataroot, verbose=True)
val = get_prediction_challenge_split("mini_val", dataroot=dataroot)

# prepare scene_name:sample_tokens map
name2sample_token = {}
scene_name2instance_token = {}
scene_instance_min_max = {}
for scene in nusc.scene:
    scene_sample_tokens = [scene['first_sample_token']]
    sample = nusc.get('sample', scene['first_sample_token'])
    instance_list = []
    instance_min_max = []
    for ann_token in sample['anns']:
        ann = nusc.get('sample_annotation', ann_token)
        if ann['instance_token'] not in instance_list:
            instance_list.append(ann['instance_token'])
            if 'vehicle' in ann['category_name'] and 'bicycle' not in ann['category_name'] and 'motorcycle' not in ann['category_name']:
                instance_min_max = [[ann['translation'][0], ann['translation'][1]]]
            
    while sample['next'] != "":
        scene_sample_tokens.append(sample['next'])
        sample = nusc.get('sample', sample['next'])
        for ann_token in sample['anns']:
            ann = nusc.get('sample_annotation', ann_token)
            if ann['instance_token'] not in instance_list:
                instance_list.append(ann['instance_token'])
                if 'vehicle' in ann['category_name'] and 'bicycle' not in ann['category_name'] and 'motorcycle' not in ann['category_name']:
                    instance_min_max.append( [ann['translation'][0], ann['translation'][1]])
     
    instance_min_max = np.array(instance_min_max)
    name2sample_token[scene['name'][-4:]] = scene_sample_tokens
    scene_name2instance_token[scene['name'][-4:]] = instance_list
    scene_instance_min_max[scene['name'][-4:]] = [np.min(instance_min_max[:,0]), 
                                                  np.max(instance_min_max[:,0]),
                                                  np.min(instance_min_max[:,1]),
                                                  np.max(instance_min_max[:,1])]
    
    #print(len(scene_sample_tokens))
#print(id2sample_token)
```

## Create submission.json

```python
def create_submission(scenes):
    ph = 12
    submission = []
    for scene in scenes:
        # scene = eval_scenes[0]
        print("processing scene: " + scene.name)
        print(f"total timesteps in scene: {scene.timesteps}")
        sample_token_list = name2sample_token[scene.name]
        instance_list = scene_name2instance_token[scene.name]
        inst_min_max = scene_instance_min_max[scene.name]

        x_min = inst_min_max[0] - 50
        y_min = inst_min_max[2] - 50
       
        with torch.no_grad():
            #timestep = 1
            for timestep in range(1, scene.timesteps):
                predictions = eval_stg.predict(scene,
                                               np.array([timestep]),
                                               ph=ph,
                                               num_samples=25)
                prediction_dict, histories_dict, futures_dict = prediction_output_to_trajectories(predictions,
                                                                                                  scene.dt,
                                                                                                  12,
                                                                                                  ph=ph,
                                                                                                  map=None)
                assert (len(prediction_dict.keys()) <= 1)
                if len(prediction_dict.keys()) == 0:
                    raise ValueError("")
                ts_key = list(prediction_dict.keys())[0]

                prediction_dict = prediction_dict[ts_key]
                histories_dict = histories_dict[ts_key]
                futures_dict = futures_dict[ts_key]

                node_list = sorted(histories_dict.keys(), key=lambda x: x.id)
                for node in node_list:
                    if node.id + "_" + sample_token_list[timestep] in val:
                        history = histories_dict[node] + np.array([x_min, y_min])
                        future = futures_dict[node] + np.array([x_min, y_min])
                        predictions = prediction_dict[node] + np.array([x_min, y_min])
                        #print(future.shape)z
                        if node.type.name == 'VEHICLE' and node.id != "ego":
                            # print(node.id in instance_list)
                            predictions_mean = np.mean(predictions, axis=1)
                            #print(node.id, np.mean(predictions, axis=1))
                            submission.append({'instance': node.id, 
                                               'sample':sample_token_list[timestep], 
                                               'probabilities':[1.0], 
                                               'prediction': predictions_mean.tolist()})

    json.dump(submission, open(submission_path, 'w'))
```

```python
create_submission(eval_scenes)
```

```python
layers = ['drivable_area',
          'road_segment',
          'lane',
          'ped_crossing',
          'walkway',
          'stop_line',
          'road_divider',
          'lane_divider']
```

## Prediction including Map Encoding

```python
ph = 12
with torch.no_grad():
    timestep = np.array([37])
    predictions = eval_stg.predict(scene,
                                   timestep,
                                   ph,
                                   num_samples=500)
    
    #print(type(list(predictions[2].keys())[0]))
    predictions_mm = eval_stg.predict(scene,
                                      timestep,
                                      ph,
                                      num_samples=1,
                                      z_mode=True,
                                      gmm_mode=True)

    # Plot predicted timestep for random scene in map
    my_patch = (x_min, y_min, x_max, y_max)
    fig, ax = nusc_map.render_map_patch(my_patch, layers, figsize=(10, 10), alpha=0.1, render_egoposes_range=False)

    ax.plot([], [], 'ko-',
            zorder=620,
            markersize=4,
            linewidth=2, alpha=0.7, label='Ours (MM)')

    ax.plot([],
            [],
            'w--o', label='Ground Truth',
            linewidth=3,
            path_effects=[pe.Stroke(linewidth=4, foreground='k'), pe.Normal()])

    plot_vehicle_nice(ax,
                      predictions,
                      scene.dt,
                      max_hl=10,
                      ph=ph,
                      map=None, x_min=x_min, y_min=y_min)
    plot_vehicle_mm(ax,
                    predictions_mm,
                    scene.dt,
                    max_hl=10,
                    ph=ph,
                    map=None, x_min=x_min, y_min=y_min)

    ax.set_ylim((1385, 1435))
    ax.set_xlim((850, 900))
    leg = ax.legend(loc='upper right', fontsize=20, frameon=True)
    ax.axis('off')
    for lh in leg.legendHandles:
        lh.set_alpha(.5)
    ax.get_legend().remove()
    fig.show()
    fig.savefig('plots/qual_nuScenes_map_pos.pdf', dpi=300, bbox_inches='tight')
```
